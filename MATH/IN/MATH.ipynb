```python
import numpy

FILE_MATH = 'Coefficients-B3.txt'
FILE = open(FILE_MATH,'r').readlines()
```

```python
MATH = ''
for LINE in FILE:

    LINE = LINE.replace('Log','numpy.log')
    LINE = LINE.replace('[','(')
    LINE = LINE.replace(']',')')
    LINE = LINE.replace(' ','')
    MATH = MATH + LINE

MATH = MATH.replace('=', ' = ')
MATH = MATH.replace('+', ' + ')
MATH = MATH.replace('-', ' - ')
MATH = MATH.replace('*', ' * ')
MATH = MATH.replace('/', ' / ')
MATH = MATH.replace('^','**')

open(FILE_MATH,'w').writelines(MATH)
```
